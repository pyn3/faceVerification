```python
import random
from faker import Faker
import pandas as pd
import numpy as np
random.seed(31)
Faker.seed(31)
fake = Faker(['th_TH'])

num_data = 1000

# create word list for random
gender = ['Male', 'Female']
gender_preferance = ['Men','Women','Non-binary','Everyone']
relationship_goal = ['Serious relationship','Someting casual','Prefer not to say','I am not sure yet']
edu_lvl = ['undergraduate','graduate','postgraduate']
province = [fake.administrative_unit() for _ in range(0,num_data)]
district = [fake.amphoe() for _ in range(0,num_data)]
ethicity = ['Asian','Arab','Africa','Native American',' Pacific Isander','South Asian','Caucasian','Other']
is_smoking = ['True','False']
is_drinking = ['True','False']
is_exercising = ['True','False']
fake = Faker(['en_TH'])
email = [fake.first_name() + '@student.xxx.ac.th' for _ in range(0,num_data)]
```

```python
# create dataframe columns
user_df = pd.DataFrame(columns=['gender','gender_preference','relationship_goal','edu_lvl','province','district','ethicity','is_smoking','is_drinking','is_exercising','email'])

# randomly create data
for i in range(0,num_data):
    user = dict()
    user['gender'] = random.choice(gender)
    user['gender_preference'] = random.choice(gender_preferance)
    user['relationship_goal'] = random.choice(relationship_goal)
    user['edu_lvl'] = random.choice(edu_lvl)
    user['province'] = random.choice(province)
    user['district'] = random.choice(district)
    user['ethicity'] = random.choice(ethicity)
    user['is_smoking'] = random.choice(is_smoking)
    user['is_drinking'] = random.choice(is_drinking)
    user['is_exercising'] = random.choice(is_exercising)
    user['email'] = random.choice(email)
    # user_df.append(user,ignore_index=True)
    user_df = pd.concat([user_df,pd.DataFrame(user, index = [0])], axis = 0)
# reset index after insertion
user_df = user_df.reset_index().drop(['index'], axis = 1)
```

```python
# display top five columns
user_df.head(5)
```

```python
# encoding binary value -> one-hot encoding
binary_col = ['gender','gender_preference','is_smoking','is_drinking','is_exercising']

# create one-hot dataframe without drop_first
one_hot_df = pd.get_dummies(user_df[binary_col])
# create one-hot dataframe with drop_first
drop_na = pd.get_dummies(user_df[binary_col], drop_first= True)
```

```python
one_hot_df.astype(float)
```

```python
drop_na.astype(float)
```

```python
user_df = user_df.drop(binary_col, axis = 1)
user_df = pd.concat([user_df,drop_na], axis = 1)
```

```python
display(user_df)
```

```python
from sklearn.preprocessing import LabelEncoder
non_bi_col = [col for col in user_df.columns if not col in binary_col]
label_encoder = LabelEncoder()
for col in non_bi_col:
    user_df[col] = label_encoder.fit_transform(user_df[col]).astype('float')
```
