```python
# root dir
digi_dir = './lfw-deepfunneled/'
# dataframe to store identity and path
ds_df = pd.DataFrame()
# loop through all folders
for root, dirs, files in os.walk(digi_dir):
    # split each deepness of the directory.
    splitted_root = root.split('/')
    # print(splitted_root)
    # loop though the deepest directory only.
    if len(splitted_root) != 3:
        continue
    # create file path for all pictures belonging to specific identity
    for file in files:
        temp_df = pd.DataFrame([(splitted_root[2],root+'\\'+file)])
        ds_df = pd.concat([ds_df,temp_df],axis = 0)
    #
    print('added '+ splitted_root[2])
    # create a temp dataframe for each identity
    # concat the temp dataframe with the path dataframe

# rename and reset index of the dataframe
ds_df = ds_df.rename({0:'identity', 1:'path'}, axis = 1).reset_index().drop('index', axis = 1)
```

```python
def strips(x):
    return x.strip()

ds_df['identity'] = ds_df['identity'].apply(strips)
ds_df[ds_df['identity'] != ''].reset_index().drop('index', axis = 1)
```

```python
ds_df.to_csv('./csv_files/lfw_ds.csv')
```
