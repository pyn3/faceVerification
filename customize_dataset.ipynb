```python
# img = Image.open('./Dataset\\Aut\\picture\\IMG_1292.JPG').convert('RGB')
# img = ImageOps.exif_transpose(img)

# face_locations = face_recognition.face_locations(np.array(img))
# for face_location in face_locations:

#     # Print the location of each face in this image
#     top, right, bottom, left = face_location
#     if top - bottom < 224:
#         print('x')

#     # You can access the actual face itself like this:
#     face_image = np.array(img)[top:bottom, left:right]
#     pil_image = Image.fromarray(face_image)
#     plt.imshow(pil_image)

def face_reg(path):
    img = Image.open(path).convert('RGB')
    img = ImageOps.exif_transpose(img)
    img = np.array(img)
    
    img_list = list()
    
    try:

        detector = MTCNN()
        face_locations = detector.detect_faces(img)
        print("I found {} face(s) in this photograph.".format(len(face_locations)))

        for face_location in face_locations:

            # Print the location of each face in this image
            x1, y1, width, height = face_location['box']

            x2, y2 = x1 + width, y1 + height
            
            face_image = np.array(img)[y1:y2, x1:x2]
            # You can access the actual face itself like this:
            pil_image = Image.fromarray(face_image)
            img_list.append(pil_image)
#             plt.imshow(pil_image)
        return (True, img_list)
    except UnboundLocalError:
        return (False, None)

```

```python
do_list = ['Pin', 'Moji','Pond']
```

```python
# break down video into frame
skip_names = []
main_dir = './Dataset'
for root, dir, files in os.walk(main_dir):
    print(root.split('\\')[-1])
    # skipping outest root
    if len(root.split('\\')) < 2 or root.split('\\')[-1] in ['video', 'picture'] or (root.split('\\')[-1] not in do_list):
        continue
    video_path = root + '\\video'
    pic_path = root + '\\picture'

    # create video directory for storing frames
    if not os.path.exists(video_path):
        os.makedirs(video_path)
        print('Make video dir')
    # create picture directory for storing picture
    if not os.path.exists(pic_path):
        os.makedirs(pic_path)
        print('Make pic dir')

    for file in files:
        file_type = file.split('.')[-1]
        file_path = root + '\\' + file
        if file_type.upper() == 'MOV':
            vidcap = cv2.VideoCapture(file_path)
            success,image = vidcap.read()
            count = 0
            print(success)
            while success:
                cv2.imwrite(video_path + "\\frame%d.jpg" % count, image)     # save frame as JPEG file
                success,image = vidcap.read()
                print('Read a new frame: ', success)
                count += 1

        if file_type.upper() in ['JPG']:
            shutil.move(file_path, pic_path + '\\' + file)
        if file_type.upper() in ['HEIC']:
            img = Image.open(file_path)
            img.save(pic_path + '\\' + file[:-4]+'jpg', format="jpeg")

```
